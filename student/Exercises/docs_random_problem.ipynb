## 1. توليد كلمة مرور

اكتب إجراءً يقبل المعاملات التالية ليُنتِج **كلمة مرور** (Password):

- الطول
- الأحرف الصغيرة
- الأحرف الكبيرة
- الأرقام
- الحد الأدنى من الأرقام
- الرموز: `!@#$%^&*`
- الحد الأدنى من الرموز

```python
def generate_password(length: int, ...) -> str:
    ...
```

````python
import string
import random
#define the allowed chars 
capital_letters=string.ascii_uppercase
lower_letters=string.ascii_lowercase
capital_lower_letters=capital_letters+lower_letters
digits=string.digits
symbols="!@#$%^&*"

def generate_password(length:int=12,
                      use_lower:bool=True,
                      use_upper:bool=True,
                      use_digits:bool=True,
                      digits_count:int=4,
                      use_symbols:bool=True,         #my first approach here was passing the numbers, chars, special in a list 
                      symbols_count:int=1) -> str: #but after asking my VA it was easier to use boolean expressions
    """this function generates a random password based
      on the rules given in the parameters"""
    if not (use_lower or use_upper or use_digits or use_symbols): 
        raise ValueError("no chars are allowed to generate a password")
    
    if digits_count + symbols_count > length:        
      raise ValueError("digits_count + symbols_count exceeds length")

    remaining_length=length

    password =""
    if use_digits:
        password+="".join(random.choices(digits,k=digits_count))
        remaining_length -= digits_count
    if use_symbols:
        password+="".join(random.choices(symbols,k=symbols_count))
        remaining_length -= symbols_count
    if use_lower and use_upper:
        password+="".join(random.choices(capital_lower_letters,k=remaining_length))
    elif use_upper:
        password+="".join(random.choices(capital_letters,k=remaining_length))
    else: #use_lower == True
        password+="".join(random.choices(lower_letters,k=remaining_length))

    #shuffle the password
    chars=list(password)
    random.shuffle(chars)
    password = "".join(chars)

    return password
random.seed(42)
print(generate_password())

    
````

## 2. توليد عبارة مرور

اكتب إجراءً آخر يقبل المعاملات التالية ليُنتِج [**عبارة مرور** (Passphrase)](https://proton.me/blog/what-is-passphrase)، بحيث تكون سهلة الحفظ على البشر:

- عدد الكلمات
- فاصل الكلمات
- تكبير الحرف الأول من كل كلمة
- تضمين رقم

```python
def generate_passphrase(...):
    ...
```

````python
#extract the python built in word list
with open("/usr/share/dict/words") as f:
      words = f.read().splitlines()
words =[word for word in words if len(word)==4]
def generate_passphrase(num_words:int=4,
                        separator:str='-',
                        capitalize_first:bool=True,
                        include_number:bool=False) -> str:
    """                                                              
  Generates a passphrase from random words.
                                                                   
  Args:           
      num_words: number of words in the passphrase                                              
      separator: separator between the words
      capitalize_first: True if you want the words to be capitalized
      include_number: include a number or not 

  Returns:
      a passphrase string
  """
    if num_words ==0:
        return ""
    words_chosen = random.choices(words,k=num_words)
    number_chosen = [random.choice(digits)]

    if capitalize_first:
         words_chosen = [word.capitalize() for word in words_chosen]
         
    if include_number:
        passphrase_list =  words_chosen + number_chosen
    else:
         passphrase_list = words_chosen

    random.shuffle(passphrase_list)

    passphrase = separator.join(passphrase_list)
    return passphrase
      
    
````

````python
random.seed(42)
print(generate_passphrase())
print(generate_passphrase(separator='_', include_number=True)) 
print(generate_passphrase(separator='_', include_number=True))   
print(generate_passphrase(separator='_', include_number=True))   
print(generate_passphrase(num_words=3, capitalize_first=False))
````
